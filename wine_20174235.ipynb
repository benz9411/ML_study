# 6.데이터 모델링
##### 기술 통계를 사용해 데이터에서 탐색한 내용으로 와인 그룹을 비교 후 품질 등급 예측 모델 

```python
#groupby 데이터 분할,데이터 연산,데이터 병합 
#column이름이나 column으로 이루어진 리스트를 전달한다.
wine.groupby('type')['quality'].describe() #groupby로 'type'부분을 분할 한 뒤 품질 열에서 describe() 실행
```

```python
wine.groupby('type')['quality'].mean() #mean함수로 평균 구하기
```

```python
wine.groupby('type')['quality'].std() #std함수로 표준편차 구하기
```

```python
wine.groupby('type')['quality'].agg(['mean', 'std']) #mean()함수랑 std()함수를 묶어서 한번에 사용하기 위해 agg()함수를 사용
```

```python
from scipy import stats  #t-검정을 위해서는 scipy 라이브러리 패키지를 사용
from statsmodels.formula.api import ols, glm #회귀 분석을 위해서는 statsmodels 라이브러리 패키지를 사용

red_wine_quality = wine.loc[wine['type'] == 'red', 'quality'] #레드 와인 샘플의 quality 값만 찾아서 red_wine에 저장
white_wine_quality = wine.loc[wine['type'] == 'white', 'quality'] #화이트 와인 샘플의 quality 값만 찾아서 white_wine에 저장 
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var = False) #scipy 패키지의 stats.ttest_ind() 함수를 사용하여 t-검정을 하고 두 그룹 간 차이를 확인
```

```python
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol' #선형 회귀 분석식의 종속 변수(y)와 독립 변수(x1~x10)를 구성
regression_result = ols(Rformula, data = wine).fit() #선형 회귀 모델 중에서 OLSOrdinary Least Squares 모델을 사용
regression_result.summary()  #선형 회귀 분석과 관련된 통계값을 출력
```

```python
sample1 = wine[wine.columns.difference(['quality', 'type'])] #wine에서 quality와 type 열은 제외하고, 회귀 분석 모델에 사용할 독립 변수만 추출하여 sample1에 저장
sample1 = sample1[0:5][:] #sample1에 있는 샘플 중에서 0번~4번 5개 샘플만 추출하고, sample1에 다시 저장하여 예측에 사용할 샘플을 제작
sample1_predict = regression_result.predict(sample1) #샘플 데이터를 회귀 분석 모델 regression_result의 예측 함수 predict()에 적용하여 수행한 뒤 결과 예측값을     sample1_predict에 저장
sample1_predict #sample1_predict를 출력하여 예측한 quality를 확인
```

```python
wine[0:5]['quality']  #wine에서 0번부터 4번까지 샘플의 quality 값을 출력하여 sample1_predict이 맞게 예측되었는지 확인
```

```python
data={"fixed_acidity" : [8.5, 8.1], "volatile_acidity":[0.8, 0.5], "citric_acid":[0.3, 0.4], "residual_sugar":[6.1, 5.8], "chlorides":[0.055,0.04], "free_sulfur_dioxide":[30.0, 31.0], "total_sulfur_dioxide":[98.0,99], "density":[0.996, 0.91], "pH":[3.25,3.01], "sulphates":[0.4, 0.35], "alcohol":[9.0, 0.88]} #회귀식에 사용한 독립 변수에 대입할 임의의 값을 딕셔너리 형태로 제작
sample2 = pd.DataFrame(data, columns=sample1.columns) #딕셔너리 형태의 값과 sample1의 열 이름만 뽑아 데이터프레임으로 묶은 sample2를 제작
sample2  #sample2를 출력
```

```python
sample2_predict = regression_result.predict(sample2) #샘플 데이터를 회귀 분석 모델 regression_result의 예측 함수 predict()에 적용하여 수행한 뒤 결과 예측값을    sample2_predict에 저장
sample2_predict  #sample2_predict를 출력하여 예측한 quality를 확인
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns #시각화에 필요한 패키지 로드
sns.set_style('dark') #히스토그램 차트의 배경색 스타일을 설정
sns.distplot(red_wine_quality, kde = True, color = "red", label = "red wine")  #레드 와인에 대한 distplot 객체를 생성
sns.distplot(white_wine_quality, kde = True, label = "white wine") #화이트 와인에 대한 distplot 객체를 생성
plt.title("Qualit of Wine Type") #차트 제목을 설정
plt.legend()  # 차트 범례를 설정
plt.show()  #설정한 내용대로 차트를 표시
```

```python
import statsmodels.api as sm //부분 회귀 계산을 위해 statsmodels.api를 로드
others = list(set(wine.columns).difference(set(["quality", "fixes acidity"])))  #부분 회귀에 사용한 독립 변수와 종속 변수를 제외한 나머지 변수 이름을 리스트 others로 추출
p, resides = sm.graphics.plot_partregress("quality","fixed_acidity", others, data = wine, ret_coords = True) #나머지 변수는 고정하고 fixed_acidity가 종속 변수 quality에 미치는 영향에 부분회귀를 수행
plt.show  #부분 회귀의 결과를 플롯으로 시각화하여 나타냄
fig = plt.figure(figsize = (8, 13))  #차트의 크기를 지정
sm.graphics.plot_partregress_grid(regression_result, fig = fig)  #다중 선형 회귀 분석 결과를 가지고 있는 regression_result를 이용해 각 독립 변수의 부분 회귀 플롯을 구함
plt.show() #부분 회귀 결과를 플롯으로 시각화하여 나타냄
```
